1. Obtengo el checkpoint.

```python
# COMPLETAR - AGREGAR TANTAS CELDAS COMO SEA NECESARIO
#Checkpoint
data2=properati[((properati.property_type=='Departamento')|(properati.property_type=='PH')|(properati.property_type=='Casa'))
               & (properati.l2=='Capital Federal')]

data2=data2[(data2.surface_total<=1000)&(data2.surface_total>=15)]
data2=data2[(data2.price<=4000000)]
data2=data2[['rooms','bedrooms','bathrooms','surface_total','surface_covered','price']]
data2.dropna(inplace=True)
data2.shape
```

2. Como métrica para analizar los modelos utilizare RMSE (Root-mean-square deviation), debido a que es la más utilizada para estos tipos de problemas y de esta manera puedo observar la amplitud del error. 

3. Las variables predictoras son `rooms`, `bedrooms`, `bathrooms`, `surface_total` y `surface_covered`. Y la variable a predecir es `price`.

```python
# Selecciono las variables predictoras
X=data2[['rooms','bedrooms','bathrooms','surface_total','surface_covered']]

y=data2['price']
```

4. Realizo la división de datos Train/Test.

```python
#Genero mi split de los datos, para entrenar y hacer la prueba
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

5. Entreno un modelo de Regresión lineal para utilizarlo como modelo de benchmark. 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as MSE

linear_model = LinearRegression()
linear_model.fit(X_train, y_train)


y_train_pred = linear_model.predict(X_train)
y_test_pred = linear_model.predict(X_test)

rmse_train = np.sqrt(MSE(y_train, y_train_pred))
rmse_test = np.sqrt(MSE(y_test, y_test_pred))

print(f'Raíz del error cuadrático medio en Train: {rmse_train}')
print(f'Raíz del error cuadrático medio en Test: {rmse_test}')
print('\n')

```
